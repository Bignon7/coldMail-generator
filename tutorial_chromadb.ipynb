```python
import chromadb

client = chromadb.Client()
collection = client.create_collection(name="my_collection")

collection.add(
    documents = [
        'This document is about Seoul',
        'This document is about Jeju Island',
        'This document is about Busan',
        'This document is about Daegu',
        'This document is about Incheon',
    ],
    ids = [
        'id1',
        'id2',
        'id3',
        'id4',
        'id5',
    ]
)
```

```python
all_docs = collection.get()
all_docs
```

```python
document = collection.get(ids=["id1"])
document
```

```python
results = collection.query(
    query_texts= ['Query is about South Korea capital'],
    n_results=2
)
results
```

```python
collection.delete(ids=all_docs['ids'])
collection.get()
```

```python
collection.add(
    documents = [
        "This document is about New York",
        "This document is about Delhi"
    ],
    ids = [
        "id3",
        "id4"
    ],
    metadatas = [
        {"url": "https://en.wikipedia.org/wiki/New_York_City"},
        {"url": "https://en.wikipedia.org/wiki/New_Deli"}
    ]
)
```

```python
results = collection.query(
    query_texts = ["Query is about Chloe Bhature"],
    n_results = 2
)
```
